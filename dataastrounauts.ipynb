```python
#library pandas
import pandas as pd   
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.cm as cm
import matplotlib as mpl
```

```python
#membaca data
df = pd.read_csv("astronauts.csv")
```

```python
df.head()
```

```python
#Perbandingan kategori

#Histogram menampilkan banyak data tahun
df["Year"].plot(kind="hist",rwidth =0.8, color="pink")
plt.title('Grafik Banyak Data Pertahun')
plt.savefig("D:/1.png")
```

```python
#Perbandingan kategori

#Bar chart menampilkan jumlah gender male dan female
df["Gender"].value_counts().plot(kind="bar", color="yellow")
plt.title('Grafik Gender')
plt.ylabel('Jumlah')
plt.xlabel('Gender')
plt.savefig("D:/2.png")
plt.show()
```

```python
#Penampilan perubahan terhadap waktu

#Line chart menampilkan jumlah pensiunan
target_df= df.Year.loc[df['Status']=='Retired']
target_df.value_counts().plot(kind="line",x="Year",y="Retired",title="Jumlah data Pensiunan", color="purple")
plt.savefig("D:/3.png")
```

```python
#Penampilan perubahan terhadap waktu

#Area chart jumlah yang almarhum dari tahun ke tahun
target_df= df.Year.loc[df['Status']=='Deceased']
target_df.value_counts().plot(kind="area",x="Year",y="Deceased",title="Jumlah data almarhum per tahun", color="red")
plt.savefig("D:/4.png")
```

```python
#Penampilan hierarki dan hubungan keseluruhan-bagian

#Komposisi banyaknya data setiap Status
df["Status"].value_counts().plot(kind = "pie")
plt.title('Grafik Data Status')
plt.savefig("D:/5.png")
```

```python
#Penampilan hierarki dan hubungan keseluruhan-bagian

#Banyaknya data per group per military rank
df.groupby (["Group","Military Rank"])["Name"].size().unstack().plot(kind="bar", stacked=True)
plt.title('Grafik Data Military Rank by Group')
plt.savefig("D:/6.png")
```

```python
#Plotting relationships

#Bubble plot menampilkan hubungan antara space flights dan space walks
df.plot(kind="scatter", x="Space Flights", y="Space Walks", color="green")
plt.title('Relationships Space Flights and Wals')
plt.savefig("D:/7.png")
```
